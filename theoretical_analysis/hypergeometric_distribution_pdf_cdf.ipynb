## pmf of hypergeometric: P(r = k) where k = [0, L]

```python
# n: dataset size
# L: number of known rows
# V: partial view size
[n, L, V] = [500000, 500, 5000]
rv = hypergeom(n, L, V)
x = np.arange(0, 30)
pmf_rows = rv.pmf(x)
fig = plot.figure()
ax = fig.add_subplot(111)
ax.plot(x, pmf_rows, 'bo')
ax.vlines(x, 0, pmf_rows, lw=2)
ax.set_xlabel('# of known records in chosen PV')
ax.set_ylabel('hypergeom PMF')
plot.title('pmf of n = 500000 L = 500 V = 5000')
# plot.savefig('P(known_records_in_PV_n_500000)', dpi=300, bbox_inches='tight')
plot.show()
```
